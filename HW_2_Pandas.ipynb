‚úçüèª **Problem 1**: Describe multi-indexing in Pandas, using a code example in your explanation. You should have a markdown cell and a code cell.

**Ans:**

**Multi-indexing:** The indices are used to accessing the rows. To locate, we use **_iloc_** and **_loc_** in pandas. Here, to index we can use multiple indices to locate. It can be understood by following examples.

```python
sd = pd.Series(["Telugu","Hindi","Tamil","Spanish","Korean","English"])
sd
```

```python
display(sd.iloc[2])
display(sd.iloc[3:6])
```

```python
sd = pd.Series(["Telugu","Hindi","Tamil","Spanish","Korean","English"],index=["a","b","c","d","e","f"])
sd
```

```python
display(sd.loc["b"])
display(sd.loc["c":"f"])
```

#### Now we can multi-index this series , it is as follows

```python
sd = pd.Series(["Telugu","Hindi","Tamil","Spanish","Korean","English"],index=[["a","b","c","d","e","f"],[0,1,2,3,4,5]])
sd
```

```python
display(sd.loc["b":"e"])
display(sd.iloc[1:4])
```

```python
import streamlit as st
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import plotly.figure_factory as ff
import plotly.express as px
df=sns.load_dataset("iris")
df2=sns.load_dataset("diamonds")

#plotly plots
data=[df["sepal_length"],df["sepal_width"],df["petal_length"],df["petal_width"]]
labels=['Sepal Length','Sepal Width','Petal Length','Petal Width']
fig1 = ff.create_distplot(data,labels,bin_size=[0.2,0.3,0.4,0.5])
fig2 = px.scatter_3d(df2,x="price",
                   y="carat",z="depth",
                   color='color')
fig3 = px.line_3d(df,x=df2["clarity"], y=df2["carat"],z=df2["color"],color=df2["cut"])
fig4 = px.violin(df, x="species", y="petal_width")

#Matplotlib plots
fig5, ax = plt.subplots()
ax.plot(df['species'], df['sepal_length'])
ax.plot(df['species'], df['sepal_width'])
ax.set_title('Sepal lengths and widths over species')
ax.set_xlabel('Species')
ax.set_ylabel('Sepal dimensions')
ax.set_ylim(bottom=0)
ax.legend()
plt.show()

st.write("""
 The Following are the some plots I have tried with plotly and Matplotlib:
 Distplot, 3D-Scatterplot, 3D-Lineplot,Violin plots,Line plots.
""")
st.plotly_chart(fig1, use_container_width=True)
st.plotly_chart(fig2, use_container_width=True)
st.plotly_chart(fig3, use_container_width=True)
st.plotly_chart(fig4, use_container_width=True)
st.pyplot(fig5, use_container_width=True)
```

‚úçüèª **Problem 5**: In a markdown cell, describe each of these concepts and how they are related (or not). 
* unstructured data,
* longitudinal data,
* structured data,
* data matrix,
* panel data,
* tidy data. 

- **Unstructured Data:**
  The Information which does not have pre-defined data. It does not have a data model. **Eg:** Text documents, Image Data, Emails etc.

- **Logitudinal Data:**
   The Information which is includes the observations of same things at different points of time. **Eg:** Population data, Employment data etc. 
   
- **Structured Data:**
   The Information which have pre-defined data. This has a data model. It is the opposite of Unstructured Data. **Eg:** Tablulated data etc.
 
- **Data Matrix:**
    It is defined as a mathematical (i.e., linear algebra) matrix containing data organized in a tidy way. **Eg:** Iris Data etc.
    
- **Panel Data:**
   Panel data is a subset of longitudinal data where observations are for the same subjects each time. **Eg:** GDP Data etc.
   
- **Tidy Data:**
   It is defined as an organization or display of data such that columns are features and rows are samples. **Eg:** Quarterly Sales Data etc.
